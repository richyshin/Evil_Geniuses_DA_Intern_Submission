# 6. Define out a posting strategy for our social channels based on your discoveries.

It was found that the engagement rate across all of the companies socials is 8.51%

If we would like to achieve an engagement rate of 15% then I would like to propose the following posting strategy.

# So when should we post?

# The days with the highest engagement were Sunday and Tuesday
The engagment rates for these two days were 10.60% and 10.16%, respectively, so posting on these two days would yield higher engagement rates.

There were two other days that showed high engagement rates as well. Friday and Thursday could also be possibe days to post with engagement rates of 8.58% and 8.97%, respectively.


# The hours with the highest engagement were hours 5, 7, and 20
These hours showed engagement rates of 26.28%, 14.34%, 13.38%, respectively.  Posting one or two posts during these hours on the above mentioned dates could lead to higher engagement rates which could lead us closer to our target.


# What types of post should be prioritized?

# The best Media Type for high engagement rates is Photos and Videos

Photos and Videos had engagement rates of 9.07% and 8.11% respectively.  Theses media types are easy for fans to interact with and share so the social media team should prioritize these more over the rest.


# What is the best way to interact with the community?

Commuity Engagement had an engagement rate of 8.59% which was double the engagement rates seen in Evergreen and Evil Exhibited.
