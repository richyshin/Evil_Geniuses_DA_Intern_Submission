# 2. Does day of the week and time of posting affect engagement rates?

```python
# import relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to tackle this task we will have to look at the average engagment rates corresponding with each time
```

```python
# first read in the dataset
pd.set_option('display.width', None) # make sure to see every column of the datasetf
df = pd.read_excel('social_data.xlsx') # read in excel file to pandas dataframe
```

```python
df.head()
```

```python
df.info()
```

```python
# the Published Date column is already a datetime object so no need to any conversions
# I would like to separate the date into Day of the Week and Hour

# create new columns for Day of the Week and Hour
# create new columns that extract the day and month from the published date
df['Day of Week'] = df['Published Date'].dt.day_name()
df['Hour of Day'] = df['Published Date'].dt.hour
df.head(3) 
```

```python
# we now have Day of the Week and Hour of Day (Assuming the times are in Pacific Time Zone)

# we now want to calculate the average engagment rate for each day and hour
```

```python
# calculate the average engagement rate for each day of the week

# group by day column and calculate the sum of total engagements and total impressions for each day
# this also creates a new dataframe that contains only the info that we want
avg_engagement_by_day = df.groupby('Day of Week')[['Total Engagements', 'Total Impressions']].sum()
# create an Engagement Rate column for each day
avg_engagement_by_day['Engagement Rate'] = (avg_engagement_by_day['Total Engagements'] / avg_engagement_by_day['Total Impressions']) * 100
# print out the dataframe
avg_engagement_by_day
```

```python
# visualize this data
# plot average engagement rates by day of the week
# Plotting average engagement rates by day of the week
plt.figure(figsize=(8, 6))
avg_engagement_by_day['Engagement Rate'].plot(kind='bar', color='blue')
plt.title('Average Engagement Rate by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Engagement Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# calculate the average engagement rate for each hour of the day

# group by Hour column and calculate the sum of total engagements and total impressions for each day
# this also creates a new dataframe that contains only the info that we want
avg_engagement_by_hour = df.groupby('Hour of Day')[['Total Engagements', 'Total Impressions']].sum()
# create an Engagement Rate column for each hour
avg_engagement_by_hour['Engagement Rate'] = (avg_engagement_by_hour['Total Engagements'] / avg_engagement_by_hour['Total Impressions']) * 100
# print out new dataframe
avg_engagement_by_hour
```

```python
# visualize the data
# Plotting average engagement rates by hour of the day
plt.figure(figsize=(8, 6))
avg_engagement_by_hour['Engagement Rate'].plot(kind='line', color='red')
plt.title('Average Engagement Rate by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Engagement Rate')
plt.xticks(range(24))
plt.show()
```

In this notebook the average engagement rate were calculated for each day of the week and the hour of the day.  From our findings it was found that Sunday and Tuesday had the highest engagement rates with 10.60% and 10.16%, respectively. Friday and Thursday coming in right behind them with 8.58% and 8.97%, respectively. 

For the hour of the day, there was exceptionally high engagement at 5 AM with a 26.28% engagement.  Assuming the time is in pacific time zone, this is 8 AM on the east coast so people who are just waking up on the East Coast could be a possible explanation for why there is such high engagement.  The next highest engagement rates were at hours 7 and 20 with engagement rates of 14.34% and 13.39%, respectively.

From these results, if we want high engagement rates, it would be most beneficial to post on Sundays or Tuesdays at 5 AM or 7 AM.  Hopefully posting posting at these dates would yield a higher engagement rate and possibly reach the target of 15%.
